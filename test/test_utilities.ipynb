```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Histogram

```python
df = pd.DataFrame({'a': np.random.randn(1000)}, columns=['a'])
ax = df.plot(kind='hist', color='skyblue', edgecolor='w', bins=20, alpha=0.7, legend=False)
axis_helper(ax)
title_helper(ax, 'Test utility functions', fontsize=16)
ax.set_ylabel('Degree',color='grey')
ax.patch.set_facecolor('#d9d9d9')
ax.patch.set_alpha(0.2)
#ax.yaxis.grid(True, color = 'grey', alpha=0.7, linestyle='--')
#ax.set_axis_bgcolor('#d9d9d9')
#ax.set_axis_alpha('0.5')
```

### Boxplot

```python
df1 = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
color1 = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Gray')
ax1 = df1.plot(kind='box', color=color1, sym='r+')
axis_helper(ax1)
title_helper(ax1, 'Test utility functions', fontsize=16)
ax1.patch.set_facecolor('#d9d9d9')
ax1.patch.set_alpha(0.2)
```

### Barplot

```python
ax5=df1[['A','B']].plot(kind='bar', edgecolor='w', color=['skyblue','gold'], alpha=0.8, legend=False)
axis_helper(ax5)
title_helper(ax5, 'Test utility functions', fontsize=16)
ax5.patch.set_facecolor('#d9d9d9')
ax5.patch.set_alpha(0.2)
legend = plt.legend(title='', loc='center left', bbox_to_anchor=(1, 0.5))
frame = legend.get_frame()
frame.set_linewidth(0.5)
#frame.set_facecolor('#d9d9d9')
#frame.set_edgecolor('grey')
frame.set_alpha(0)
for text in legend.get_texts():
    text.set_color('grey')
#frame.set_labelcolor('grey')
```

### Areaplot

```python
ax2 = df1.plot(kind='area', alpha=0.4)
axis_helper(ax2, top=True, right=True)
title_helper(ax2, 'Test utility functions', fontsize=16)
ax2.patch.set_facecolor('#d9d9d9')
ax2.patch.set_alpha(0.2)
#legend = plt.legend(title='', loc='center left', bbox_to_anchor=(1, 0.5))
```

### Scatter

```python
df2 = pd.DataFrame(np.random.rand(50, 4), columns=['A', 'B', 'C', 'D'])
ax3 = df2.plot(kind='scatter', x='A', y='B', s=df2['D']*200, color='skyblue')
ax3.patch.set_facecolor('#d9d9d9')
ax3.patch.set_alpha(0.2)
axis_helper(ax3, top=True, right=True)
title_helper(ax3, 'Test utility functions', fontsize=16)
ax3.grid(True, color = 'w', alpha=0.8, linestyle='-', lw=0.5)
```

### Hex

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df3['b'] = df3['b'] + np.arange(1000)
ax4 = df3.plot(kind='hexbin', x='a', y='b', gridsize=25)
ax4.patch.set_facecolor('#d9d9d9')
ax4.patch.set_alpha(0.2)
axis_helper(ax4)
title_helper(ax4, 'Test utility functions', fontsize=16)
#ax4.grid(True, color = 'w', alpha=0.8, linestyle='-', lw=0.5)
```
